```python
## Ejercicios
import pandas as pd
import numpy as np

# 1. Creación de DataFrames
# (a) Crea un DataFrame df1 con una columna llamada numeros que contenga los valores del 0 al 19.
data = list(range(20))
df1 = pd.DataFrame({'Numeros': data})
df1

```

```python
# (b) Crea un DataFrame df2 con una columna aleatorios de tamaño 10 con números aleatorios entre 0 y 1.

np.random.seed(10)

numeros_aleatorios = np.random.rand(10)
df2 = pd.DataFrame({'Aleatorios': numeros_aleatorios})
df2

```

```python
# (c) Combina los DataFrames anteriores df1 y df2 en un nuevo DataFrame llamado df3.

df3 = pd.concat([df1,df2])
df3

```

```python
# (d) Guarda el DataFrame df3 en un fichero .csv.

df3.to_csv('mi_dataframe_3.csv', index=False)
```

```python
# 2. Indexación y selección (10 pts)
# Dado el DataFrame:
import pandas as pd
df = pd.DataFrame({
"A": range(1, 11),
"B": [5,3,6,8,2,9,1,7,4,10]
})

# (a) Selecciona las filas donde el valor de la columna B sea mayor que 5.

mascara = df['B'] > 5

df_filtrado = df[mascara]

print("### DataFrame Filtrado (B > 5) ###")
print(df_filtrado)

```

```python
# (b) Cambia los valores menores que 4 en la columna B a NaN. Para usar el valor NaN hay que utilizar el código pd.NA

mascara_nan = df['B'] < 4


df.loc[mascara_nan, 'B'] = pd.NA

print("### DataFrame con pd.NA en Columna B (B < 4) ###")
print(df)

```
