1. fixed acidity (acidez fija)

Son los ácidos que no se evaporan fácilmente (como el ácido tartárico).
Afectan el sabor, la frescura y la estructura del vino.

2. volatile acidity (acidez volátil)

Indica la cantidad de ácidos que sí se evaporan (como el ácido acético).
En exceso puede dar un olor desagradable parecido al vinagre.

3. citric acid (ácido cítrico)

Pequeña cantidad presente de forma natural.
Aporta frescura y un toque cítrico al vino.

4. residual sugar (azúcar residual)

Cantidad de azúcar que queda tras la fermentación.
Valores altos → vino más dulce.

5. chlorides (cloruros)

Representa la cantidad de sal en el vino.
Niveles altos afectan negativamente el sabor.

6. free sulfur dioxide (SO₂ libre)

Parte del dióxido de azufre que actúa como conservante y evita el crecimiento de bacterias.
Protege al vino de la oxidación.

7. total sulfur dioxide (SO₂ total)

Incluye el SO₂ libre y el combinado.
Niveles muy altos pueden afectar el aroma.

8. density (densidad)

Qué tan “pesado” es el vino comparado con el agua.
Depende del azúcar, alcohol y otras sustancias.

9. pH

Mide la acidez total del vino.
Valores bajos → más ácido.
Valores altos → menos ácido.

10. sulphates (sulfatos)

Otro tipo de conservante.
Ayuda a combatir microbios y mejorar la estabilidad del vino.

11. alcohol

Porcentaje de alcohol del vino.
Influye en el cuerpo, sabor y percepción del vino.

12. quality (calidad)

Puntuación otorgada por expertos (generalmente entre 0 y 10).

```python
# 4. Combinando información en pandas
# Dispones de los siguientes DataFrames:
import pandas as pd

estudiantes = pd.DataFrame ({
    "id": [1, 2, 3, 4],
    "nombre": ["Ana", "Luis", "Marta", "Jorge"]
})
notas = pd.DataFrame({
    "id": [1, 2, 2, 5],
    "asignatura": ["Matemáticas", "Historia", "Química", "Biología"],
    "nota": [8.5, 7.0, 6.5, 9.0]
})
print("DataFrame estudiantes:\n", estudiantes)
print("\nDataFrame notas:\n", notas)
```

```python
# Realiza las siguientes tareas:
# (a) Realiza un merge interno (inner ) entre ambos DataFrames. ¿Cuántas filas se obtienen? ¿Por qué?

merge_interno = pd.merge(estudiantes, notas, on='id', how='inner')

print("\n(a) Merge Interno:\n", merge_interno)

# Solo los IDs 1 y 2 existen en ambas tablas.

# El ID 2 (Luis) aparece dos veces en la tabla notas (Historia y Química), por lo que el resultado incluye dos filas para Luis.

# Los estudiantes con IDs 3 (Marta) y 4 (Jorge), y la nota con ID 5 se eliminan porque no tienen coincidencia en el otro DataFrame.
```

```python
# (b) Realiza un merge externo (outer ). Explica las filas que aparecen con valores nulos.

merge_externo = pd.merge(estudiantes, notas, on='id', how='outer')

print("\n(b) Merge Externo:\n", merge_externo)

# Filas 3 y 4 (Marta y Jorge): Aparecen con NaN en las columnas asignatura y nota porque no existe ningún registro de notas con el id 3 o 4 en la tabla notas.
# Fila 5 (ID 5 / Biología): Aparece con NaN en la columna nombre porque no existe un estudiante con el id 5 en la tabla estudiantes.
```

```python
# (c) Realiza un left merge usando estudiantes como tabla izquierda. ¿Qué información se pierde respecto al outer?
merge_izquierdo = pd.merge(estudiantes, notas, on='id', how='left')

print("\n(c) Merge Izquierdo:\n", merge_izquierdo)

# Información que se pierde respecto al outer merge: Se pierde la fila con el ID 5 (la nota de Biología). Esto se debe a que la 
# operación left exige que el id exista en la tabla izquierda (estudiantes), y el ID 5 solo existía en la tabla notas.
```

```python
# (d) Realiza un right merge usando notas como tabla derecha. ¿Qué estudiantes desaparecen en el resultado?

merge_derecho = pd.merge(estudiantes, notas, on='id', how='right')

print("\n(d) Merge Derecho:\n", merge_derecho)
# Estudiantes que desaparecen en el resultado: Desaparecen Marta (ID 3) y Jorge (ID 4). Esto ocurre porque no tienen ningún registro de notas 
# en la tabla derecha (notas), y la operación right solo garantiza la inclusión de las filas de la tabla derecha.
```

```python
# (e) Añade una columna llamada curso al DataFrame estudiantes. Después, combina los datos utilizando un merge múltiple 
# por id y explica el resultado.

estudiantes['curso'] = ['A', 'A', 'B', 'B']

print("\nEstudiantes con 'curso' añadido:\n", estudiantes)

# Combina los DataFrames (merge múltiple por 'id')
merge_final = pd.merge(estudiantes, notas, on='id', how='inner')

print("\n(e) Merge Múltiple Final:\n", merge_final)
```

```python
# 5. Agregaciones y operaciones con groupby
#En este ejercicio se trabajará con un conjunto de datos que contiene información sobre ventas realizadas por una empresa en distintas regiones.
# Considera el siguiente DataFrame:
import pandas as pd

ventas = pd.DataFrame({
"region": ["Norte","Norte","Sur","Este","Sur","Este","Oeste",
"Oeste","Norte","Sur"],
"vendedor": ["Ana","Luis","Ana","Jorge","Luis","Ana","Marta",
"Luis","Ana","Jorge"],
"ventas": [1200, 800, 950, 1100, 400, 1600, 700, 450, 895, 900],
"producto": ["A","A","B","A","B","A","B","A","B","B"]
})

# Realiza las siguientes tareas:
# (a) Obtén el total de ventas por región.

ventas_por_region = ventas.groupby('region')['ventas'].sum()

print("\n(a) Total de ventas por región:\n", ventas_por_region)
```

```python
# (b) Calcula la media, el valor máximo y el mínimo de las ventas agrupadas por vendedor.

resumen_por_vendedor = ventas.groupby('vendedor')['ventas'].agg(['mean', 'max', 'min'])

print("\n(b) Resumen de ventas por vendedor:\n", resumen_por_vendedor)
```
